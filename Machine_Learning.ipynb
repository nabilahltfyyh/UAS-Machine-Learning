# **Segmentasi Data Konsumen**

```python
# PENGELOMPOKAN DATA
# PENENTUAN JUMLAH CLUSTER DENGAN METODE ELBOW

x1 = df.loc[:, ['Age', 'Spending Score (1-100)']].values

from sklearn.cluster import KMeans
wcss = []

for k in range (1,11):
  kmeans = KMeans(n_clusters=k, init="k-means++")
  kmeans.fit(x1)
  wcss.append(kmeans.inertia_)

plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11), wcss, linewidth=2, color="red", marker="8")
plt.xlabel("K Values")
plt.ylabel("WCSS")
plt.show()
```

```python
# PENGELOMPOKAN DATA DENGAN K-MEANS

kmeans = KMeans(n_clusters=4)
label = kmeans.fit_predict(x1)

print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(x1[:,0], x1[:,1], c=kmeans.labels_, cmap="rainbow")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color="black", marker="*", s=200)

plt.title("Pengelompokan Customer")
plt.xlabel("Age")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

```python
# PENGELOMPOKAN DATA 
# PENENTUAN JUMLAH CLUSTER DENGAN METODE ELBOW

x2 = df.loc[:, ['Annual Income (k$)', 'Spending Score (1-100)']].values

from sklearn.cluster import KMeans
wcss = []

for k in range (1,11):
  kmeans = KMeans(n_clusters=k, init="k-means++")
  kmeans.fit(x2)
  wcss.append(kmeans.inertia_)

plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11), wcss, linewidth=2, color="red", marker="8")
plt.xlabel("K Values")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5)
label = kmeans.fit_predict(x2)

print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(x2[:,0], x2[:,1], c=kmeans.labels_, cmap="rainbow")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color="black", marker="*", s=200)

plt.title("Pengelompokan Customer")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

```python
# PENGELOMPOKAN DATA BERDASARKAN AGE, ANNUAL INCOME, DAN SPENDING SCORE

x3 = df.iloc[:,1:]

from sklearn.cluster import KMeans

wcss = []

for k in range (1,11):
  kmeans = KMeans(n_clusters=k, init="k-means++")
  kmeans.fit(x3)
  wcss.append(kmeans.inertia_)

plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11), wcss, linewidth=2, color="red", marker="8")
plt.xlabel("K Values")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5)
label = kmeans.fit_predict(x3)

print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
clusters = kmeans.fit_predict(x3)
df["label"] = clusters

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df.Age[df.label==0], df["Annual Income (k$)"][df.label==0], df["Spending Score (1-100)"][df.label==0], c='blue', s=60)
ax.scatter(df.Age[df.label==1], df["Annual Income (k$)"][df.label==1], df["Spending Score (1-100)"][df.label==1], c='red', s=60)
ax.scatter(df.Age[df.label==2], df["Annual Income (k$)"][df.label==2], df["Spending Score (1-100)"][df.label==2], c='green', s=60)
ax.scatter(df.Age[df.label==3], df["Annual Income (k$)"][df.label==3], df["Spending Score (1-100)"][df.label==3], c='orange', s=60)
ax.scatter(df.Age[df.label==4], df["Annual Income (k$)"][df.label==4], df["Spending Score (1-100)"][df.label==4], c='cyan', s=60)
ax.view_init(30,185)

plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
ax.set_zlabel("Spending Score (1-100)")
plt.show()
```
